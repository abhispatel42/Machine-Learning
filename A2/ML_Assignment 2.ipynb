```python
import numpy as np
import pylab
from scipy.stats import norm
import matplotlib.pyplot as plt
from collections import Counter

vector4= []
n = 1
while n<=20:
    num1 = np.random.randint(1,7)        #https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.randint.html
    num2 = np.random.randint(1,7)
    num3 = np.random.randint(1,7)
    num4 = np.random.randint(1,7)
    num5 = np.random.randint(1,7)
    num6 = np.random.randint(1,7)
    num7 = np.random.randint(1,7)
    
    vector4.append(num1+num2+num3+num4+num5+num6+num7)
    n= n+1


mean =np.mean(vector4)
std1=np.std(vector4)
print('The mean is: '+str(mean))
print('The standard deviation is: ' +str(std1))
plt.hist(vector4)
x = np.linspace(mean-2*std1,mean+2*std1, 100)    #https://stackoverflow.com/questions/20011122/fitting-a-normal-distribution-to-1d-data
p = norm.pdf(x, mean, std1)
plt.plot(x, p, 'k', linewidth=2)
plt.show()
```

```python
from scipy.integrate import quad
import math

def make_gauss(N, sigma, mu):      #https://stackoverflow.com/questions/509994/best-way-to-write-a-python-function-that-integrates-a-gaussian
    k = N / (sigma * math.sqrt(2*math.pi))
    s = -1.0 / (2 * sigma * sigma)
    def f(x):
        return k * math.exp(s * (x - mu)*(x - mu))
    return f

I1= quad(make_gauss(1,1, 1), -np.inf, 2) #https://docs.scipy.org/doc/scipy/reference/tutorial/integrate.html
I2= quad(make_gauss(1,3, 2), -2, np.inf)
result= (I1[0]+I2[0])/2
print('The accuracy of the model is:' +str(result))
```
